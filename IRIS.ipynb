```
temp = dataset.drop('Id', axis=1)
g = sns.pairplot(temp, hue='Species', markers='+')
plt.show()
```

```
X = dataset.drop(['Species'], axis=1)
y = dataset['Species']
```

```
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X, y)
y_pred = knn.predict(X)
```

```
metrics.accuracy_score(y, y_pred)
```
